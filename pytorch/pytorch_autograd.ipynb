autograd on NN


```python
import torch

x=torch.tensor(5.6) #input feature
y=torch.tensor(0.0) #output (binary)

w=torch.tensor(1.0) #weights
b=torch.tensor(0.0) #bias
```

```python
#bianry cross entropy loss function
def binary_cross_entropy_loss(pred, target):
  epsilon=1e-8
  pred=torch.clamp(pred,epsilon,1-epsilon)
  return -(target*torch.log(pred)+(1-target)*torch.log(1-pred))
```

```python
#forward pass
z=w*x+b
y_pred= torch.sigmoid(z)

loss=binary_cross_entropy_loss(y_pred,y)
```

```python
#derivative
dloss_dy_pred=(y_pred-y)/(y_pred*(1-y_pred))

dy_pred_dz=y_pred*(1-y_pred)

dz_dw=x
dz_db=1


dL_dw=dloss_dy_pred * dy_pred_dz * dz_dw
dL_db=dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"manual gradient loss  wrt wright(dw) : {dL_dw}")
print(f"manual gradient loss  wrt wright(dw) : {dL_db}")
```
