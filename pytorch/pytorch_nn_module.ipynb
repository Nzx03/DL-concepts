```python
import torch
import torch.nn as nn #handling the autograd functionaility internally

class Model(nn.Module):   #inherit from nn module , otherwise it wont be able to access nnn module'sm offering


    def __init__(self, num_features):
        #single neuron
       super().__init__()  #invoking parent class constructor using super
       self.linear=nn.Linear(num_features,1)  #(input,output)  ,layer=linear
       self.sigmoid=nn.Sigmoid()


    #forward pass
    def forward(self, features):

      out=self.linear(features)  #wx+b
      out=self.sigmoid(out)

      return out


```

```python
#creating dataset

features= torch.rand(10,5)  #10 rows, 5 columsn( 5 features)

#creating model
model=Model(features.shape[1])  #dimension =1 (rows)

#calling mdel
model(features)   #can me model.foward(features)  , calling as a fucntion
```

```python
#now with hidden layer
class Model1(nn.Module):   #inherit from nn module , otherwise it wont be able to access nnn module'sm offering


    def __init__(self, num_features):

       super().__init__()  #invoking parent class constructor using super
       self.linear1=nn.Linear(num_features,3)  #getting 5 input , and output getting 3 weights
       self.relu=nn.ReLU()
       self.linear2=nn.Linear(3,1)
       self.sigmoid=nn.Sigmoid()

    #forward pass
    def forward(self, features):

      out=self.linear1(features)  #wx+b
      out=self.relu(out)
      out=self.linear2(out)
      out=self.sigmoid(out)

      return out
```

```python
#creating dataset

features= torch.rand(10,5)  #10 rows, 5 columsn( 5 features)

#creating model
model=Model(features.shape[1])  #dimension =1 (rows)

#calling mdel
model(features)   #can me model.foward(features)  , calling as a fucntion
```

```python
#now with hidden layer ( to minimize this using sequnetila containers)
class Model1(nn.Module):   #inherit from nn module , otherwise it wont be able to access nnn module'sm offering


    def __init__(self, num_features):

       super().__init__()  #invoking parent class constructor using super
       self.network=nn.Sequential(
           nn.Linear(num_features,3),
           nn.ReLU(),
           nn.Linear(3,1),
           nn.Sigmoid(),
         )

    #forward pass
    def forward(self, features):

      out=self.network(features)

      return out
```

```python
#creating dataset

features= torch.rand(10,5)  #10 rows, 5 columsn( 5 features)

#creating model
model=Model(features.shape[1])  #dimension =1 (rows)

#calling mdel
model(features)   #can me model.foward(features)  , calling as a fucntion
```
